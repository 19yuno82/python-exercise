### 문자열 인덱싱 ,슬라이싱
- 인덱스 : 데이터가 생성될 때 자동으로 부여되는 순서 번호
- 인덱싱 : 무엇(인덱스)인가를 '가리킨다'는 의미
- 슬라이싱 : 무엇(인덱스)인가를 '잘라낸다'는 의미

```python
# 문자열 인덱싱
str1 = s1 + s2
print(str1)
print(str1[0])
print(str1[8])
print(str1[-1])
print(str1[-2])
```

```python
# 문자열 슬라이싱
# 변수명[시작값:끝값] ※ 시작값은 포함, 끝값은 미포함
print(str1[0:1])
print(str1[0:2])
print(str1[3:7])
```

```python
print(str1[8:15])
print(str1[8:])
print(str1[-4:])
```

```python
print(str1[0:15])
print(str1[:])
```

```python
day = "2020년 3월 3일의 날씨는 맑음입니다."
print("날짜:",day[0:11])
print("날씨 :",day[-6:-4])
```

```python
# 문자열 포메팅(formattingt)
# 문자열 안의 특정한 값을 삽입해야 할 때 사용

#첫번째 방법 : % 기호 포메팅
day = 27
month = 9
s = "오늘은 %d월 %d일 입니다."%(month,day)
print(s)
```

- 문자열 포맷코드


```python
# 두번째 방법 : format()를 사용한 포메팅
day = 27
month = 9
s = "오늘은 {}월 {}일 입니다.".format(month, day)
print(s)
```

```python
#세번째 방법 ; f문자열 포메팅
day = 27
month = 9
s = f"오늘은 {month}월 {day}일 입니다."
s
```

| 함수            | 설명                                      |
|-----------------|-------------------------------------------|
| `len()`         | 문자열의 문자 개수 반환                   |
| `upper()`       | 문자열의 소문자를 대문자로 바꿈           |
| `lower()`       | 문자열의 대문자를 소문자로 바꿈           |
| `isupper()`     | 문자열이 대문자인지 여부 반환             |
| `islower()`     | 문자열이 소문자인지 여부 반환             |
| `isdigit()`     | 문자열이 숫자인지 여부 반환               |
| `count('문자열')` | '찾을 문자열'이 몇 개 있는지 개수 반환    |
| `find('문자열')`  | '찾을 문자열' 위치 반환                   |
| `index('문자열')` | '찾을 문자열' 위치 반환                   |
| `strip()`       | 문자열의 양쪽 공백 제거                  |
| `lstrip()`      | 문자열의 왼쪽 공백 제거                  |
| `rstrip()`      | 문자열의 오른쪽 공백 제거                |
| `replace('문자1', '문자2')` | '문자1'을 '문자2'로 바꾸기       |
| `split()`       | 문자열 나누기                             |
