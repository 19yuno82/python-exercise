```python
#파일 불러오기
import pickle
```

```python
with open('./movie_review.pkl','rb') as f :
    data = pickle.load(f)
```

```python
#1. 전체 리뷰 개수 확인
data
```

```python
#2.리뷰 길이 평균 값
reLength = 0
for i in range(len(data)) :
    reLength += len(data[i][0])
print(f"리뷰의 평균 길이는 '{reLength/len(data)}'입니다.")

total = 0
for re in data :
    total += len(re[0])
print(f"리뷰의 평균 길이는 '{total/len(data)}'입니다.")
```

```python
#3.긍정리뷰/부정리뷰 개수
posSum = denSum = 0
for i in range(len(data)) :
    if data[i][1] == 1 :
        posSum += 1
    elif data[i][1] == 0 :
        denSum += 1
print(f"긍정:{posSum} 부정:{denSum}")
```

```python
#4.리뷰 길이 최대값/리뷰 내용
reLenMax = 0
for i in range(len(data)) :
    if len(data[i][0]) > reLenMax :
        reLenMax = len(data[i][0])
        s= data[i][0]
print(f"리뷰 최대 길이는 {reLenMax}입니다.")
print(s)

reMax = len(data[0][0])
for re in data :
    if len(re[0]) >reMax :
        reMax = len(re[0])
        s= re[0]
print(f"리뷰 최대 길이는 {reMax} 입니다.")
print(s)
```

```python
#5."ㅋ"이  포함된 리뷰 개수
reSum = 0
for i in range(len(data)) :
    if data[i][0].count('ㅋ') != 0 :
        reSum += 1
print(f" 'ㅋ' 이 포함된 리뷰는 {reSum}개 입니다.")
```

```python
#6.'검색데이터(입력값)'가 포함된 리뷰 개수
s = input("검색데이터 >> ")
reSum = 0
for re in data :
    if s in re[0] :
        reSum += 1
print(f" '{s}' 이 포함된 리뷰는 {reSum}개 입니다.")
```
